    A3: The Echo-Chamber Effect
    We characterise the "echo-chamber" effect as the situation where a probability is stated then a Receiver then becomes a Source for another Reciever, but then the Reciever then becomes a Source for the orginal Receiver. If the participants realised that it was their own opinion being fed back to them they would update their beliefs only to the degree to which they believe that the Source has some other independent information. The "echo-chamber" is created by the mistaken treatment of the "echo" as being independent information rather than their own opinion being fed back to them, albeit modifiied by the Source.
    Our model is as follows:
    a) There are two identical "players"
    b) A probability (p0) is fed into the system exogenously as a stated probability
    b) One of the players (a first receiver) updates their beliefs given this probability based on their beliefs about the Source
    c) The first receiver then becomes a source for the other player (the second receiver) who updates their beliefs based on the stated probability of the first receiver.
    d) The system then interates between b) and c) for a given number of iterations.

```python
from bayesprob import *
```

```python
def echo_chamber(aa,ar):
    V0,U0=[],[]
    ps=[.1,.25,.4,.45,.49, 0.5,.51,.55,.6,.75,.9] #The range probabilities for the simulation, each serve as p0 described above
    for p in ps:
        a=bayes(a_0=aa,a_r=ar)  #Player a
        b=bayes(a_0=aa,a_r=ar)  #Player b

        #The echo chamber is now simulated
        v,u=[],[]
        for i in range(100):
            a.p=p
            v+=[a.prob()]
            p=a.prob()
            a.n_r+=a.N()
            a.a_r=a.prob()
            b.p=p
            u+=[a.prob()]
            p=b.prob()
            b.n_r+=b.N()
            b.a_r=b.prob()
        U0+=[u]
        V0+=[v]

    probs=[]
    for p in ps:
        probs+=['p0='+str((p))]
    Vs=frame(V0).T
    Us=frame(U0).T
    Us.columns=probs
    Vs.columns=probs
    Vs.index.name='iterations'
    Vs.columns.name='initial probability'
    return Vs
```

# Figure B3.1

```python
pltsize(4,10)
aa=0.5  #Set the parameters of the agents, these are for a_0
ar=0.5  #Set the parameters of the agents , these are for a_r
Vs=echo_chamber(aa,ar)
tit='Eco Chamber for 100 interations, Intial Priors ' + r'$a_r$' + '=' +str(ar)+ ', ' +   r'$a_0$' + '=' +str(aa)
Vs.plot(grid=True, title=tit)
#plt.savefig(direcout+'FigA3.1'+'.png',bbox_inches='tight')
```
