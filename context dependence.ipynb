The following generates results using the bayes object. Does not deliver any specific results in the paper, but might be useful to gain a better understanding of the model


```python
from bayesprob import *
```

```python
#A simple probability object
#Below the receiver has a fixed belief that the priors of the source are fixed and the same as theirs (theta=0,phi=1)
#In this circumstance any probability will be accepted verbatim
a=bayes(p=0.01,n_0=2,a_0=.5,lamda_0=.5,theta=.5,phi=.5,delta=[.5,0],n_r=2,a_r=.5)
b,c,d=copy(a),copy(a),copy(a)
b.a_r,b.a_0=0.1,.1
c.a_r,c.a_0=0.01,.01
d.a_r,d.a_0=0.001,.001
```

```python
#The ex_ante priors and the parameters are as set above
init=pd.concat([b.initial(),c.initial(),d.initial()])
init.index=['£10','£100','£1000']
l=list(init.columns)
l[0]='Winning Prob'
init.columns=l
init=init.drop([l[0]],axis=1)
init['Prior E-Val']=[1.00,1,1]

init.index.name='Payoff'
init
```

```python
expost=pd.concat([b.expost(),c.expost(),d.expost()])
l=list(expost.columns)
expost.index=['£10','£100','£1000']
expost['Sub Expectation']=expost[l[6]]*np.array([10,100,1000])
expost['Ob Expectation']=[10/100,100/100,1000/100]
expost=expost.drop(l[8],axis=1)
expost.round(3)
```

```python
init.to_csv('tab1.csv')
```
