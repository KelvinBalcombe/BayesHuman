The following generates graphs for key parameters over a range of state probabilities. These can be set within fg below

```python
from bayesprob import *

def mapV(V):
    nam=V.columns
    St=V.iloc[950,:]
    pi_s=0.95

    phat=round(float(St[nam[6]]),3)
    nhat=round(float(St[nam[2]]),3)
    xn=  round(float(St[nam[4]]),3)

    nam=V.columns
    Vp=V[[nam[4],nam[6]]]
    Vp.columns=[nam[4]+'='+  str(xn),nam[6]+'='+ str(phat)]
    
    fig,ax=plt.subplots(1,2)

    ax[0].vlines(pi_s,0,phat, linestyles="dashed")
    ax[0].hlines(phat,0,pi_s, linestyles="dashed")
    ax[0].vlines(pi_s,0,xn, linestyles="dashed",color='black')  
    ax[0].hlines(xn,0,pi_s, linestyles="dashed",color='green')


    ax[1].vlines(pi_s,0,nhat, linestyles="dashed")
    ax[1].hlines(nhat,0,pi_s, linestyles="dashed")
   

    Vn=V[[nam[2]]]
    Vn.columns=[nam[2]+'='+str(nhat)]
    return Vn,Vp,fig,ax
```

# Figure 3.1

```python
pltsize(4,10)
fg={'lamda_0':.5,'a_0':0.5, 'n_0':2 ,'theta':0,'delta':[0.5,0],'phi':1  ,'a_r':0.5 ,'n_r':2}
V=bayesprob(fg)
Vn,Vp,fig,ax=mapV(V)
Vp.plot(grid=True,xlabel=r'$π_{s}$',fontsize=12,ax=ax[0],xlim=(.5,1),ylim=(.4,1))
Vn.plot(grid=True,xlabel=r'$π_{s}$',fontsize=10,ax=ax[1],xlim=(.8,1),ylim=(0,100))
settings=labels(fg)
fig.suptitle(settings, y=1.1, fontsize=16  )

#plt.savefig(direcout+'Fig3.1'+'.png',bbox_inches='tight')
```
