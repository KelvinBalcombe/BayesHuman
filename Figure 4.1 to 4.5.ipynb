The following generates graphs for key parameters over a range of state probabilities. These can be set within fg below

```python
from bayesprob import *
```

# Figure 4.1

```python
import matplotlib.pylab as pylab; params = {'axes.labelsize': 'x-large'};pylab.rcParams.update(params)

#Set the prior parameters here
fg={'lamda_0':.5,'a_0':0.5, 'n_0':2 ,'theta':0.5,'delta':[.5,.0],'phi':.5  ,'a_r':0.5 ,'n_r':2}

#This produces a dataframe (V) indexed by the stated probability
#Note that pie_s here will be the stated probability, but will be pie^bar_s in the paper if there is dishonesty

V=bayesprob(fg)
pltsize(4,20);
fig,ax=plt.subplots(1,5)
graphout(V,fg,fig,ax)
#plt.savefig(direcout+'Fig4.1'+'.png',bbox_inches='tight')
```
