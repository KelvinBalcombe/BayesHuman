The following generates results using the bayes object. Does not deliver any specific results in the paper, but might be useful to gain a better understanding of the model


```python
from bayesprob import *
```

# Example 1 the "base case":  
Honest source with uniform priors and equal flexibility in the parameters

```python
a_0=1/2       #The prior mean of the Source
n_0=2         #The prior concentration of the Source
a_r=1/2       #The prior mean of the Receiver
n_r=2         #The prior concentration of the Receiver
delta=[.5,.9]  #Source Honesty, [0.5,0] an honest Source
lamda_0=0.5    #Proportion of the n_0/(n_0+N_0) which implicity defines N_0
theta= 1/2    # Flexibility of lambda to a, theta=1 indicates dogmatic about lambda=lambda_0, theta=0 dogmatic about a=a_0
phi=1/2       #Flexibility of n relative to N, phi=1 indcates dogmatic view that n=n_0, phi=1 indicates dogmatic view that N=N_0


a=bayes(n_0=n_0,a_0=a_0,n_r=n_r,a_r=a_r,lamda_0=lamda_0,theta=theta,phi=phi,delta=delta)
```

```python
#the stated probability
a.p=0.1
```

```python
#The ex_ante priors and the parameters are as set above
a.initial()
```

```python
#A dataframe summary of the expost parameters are
a.expost().round(6).T
```

```python
pltsize(4,10)
a.graphs()
```

```python
#The density of log-odds
a.graphs_lodds()
```
